```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
import datetime

pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)

pytrends = TrendReq(hl='en-US', tz=360)
kw_list = ["rural electrification corporation"]
pytrends.build_payload(kw_list, cat=0, timeframe='all', geo='IN', gprop='')

graphdf = pytrends.interest_over_time()
graphdf.plot(figsize=(20, 12), y=kw_list, kind ='line')
```

```python
tabledf = pd.DataFrame(graphdf)
tabledf=tabledf.drop(tabledf.columns[1], axis=1)

print(tabledf)
```

```python
tabledf.to_csv('rec.csv')
```
